```python
import requests
googlekey = 'AIzaSyBQW2Urw2sIAWxK81wqSgfhBslTW-cgR1k'
directions_url = "https://maps.googleapis.com/maps/api/directions/json"

def getDirections(start,finish):
    
    params = {
        'destination' : finish,
        'origin' : start,
        'key': googlekey
    }
    response = requests.get(directions_url, params = params)
    response.raise_for_status()#throw python error, when HTTP error
    result = response.json() #de-seralize

    steps = result['routes'][0]['legs'][0]['steps']    
    
    return steps


route = getDirections("Stong Museum in Rochester, NY", "JMA Wireless Dome")
```

```python
## Applying a function to every row in the dataframe
def boss_sale_price(price):
    if price >100:
        return "High"
    elif price > 25:
        return "Mid"
    else:
        return "Low"
    
boss_sale_price(150)
```

```python
# apply a function call to each row in the DF axis = 1 is row
can_sale_copy.loc[:,'Boss Category'] \
    = can_sale_copy.apply(lambda row: boss_sale_price(row['Sale Price']), axis=1)
```
