## Step 1b: Code the API: Proof of Concept

```python
# PROMPT 3 Demonstrate you can call the API 
start = "Manley Field House"
finish = "JMA Wireless Dome"
googlekey = 'AIzaSyBQW2Urw2sIAWxK81wqSgfhBslTW-cgR1k'
directions_url = "https://maps.googleapis.com/maps/api/directions/json"
params = {
    'destination': finish,
    'origin' : start,
    'key' : googlekey
}
response = requests.get(directions_url, params = params)
response.raise_for_status() # takes an HTTP error and throws an exception
results = response.json() # de-serialize

results
```

```python
steps = results['routes'][0]['legs'][0]['steps']
```

```python
# PROMPT 4 What do I need from the response?
from IPython.display import display, HTML
display(HTML("<H1>This is the title!</h1>"))
for step in steps:
    display(HTML(step['html_instructions']))
```

## Step 1.c: Refactor into a function `getDirections()`

Now the we understand the API, let's refactor the code into a useable function

```python
# PROMPT 9 - write function
def getDirections(start,finish):
       
    googlekey = 'AIzaSyBQW2Urw2sIAWxK81wqSgfhBslTW-cgR1k'
    directions_url = "https://maps.googleapis.com/maps/api/directions/json"
    params = {
        'destination': finish,
        'origin' : start,
        'key' : googlekey
    }
    response = requests.get(directions_url, params = params)
    response.raise_for_status() # takes an HTTP error and throws an exception
    results = response.json() # de-serialize

    steps = results['routes'][0]['legs'][0]['steps']
    
    return steps
```

## Step 1.d: Test the function

```python
# PROMPT 10
from IPython.display import display, HTML

start = "8031 Capri Circle Clay, NY 10341"
end = "JMA Wireless Dome"
steps = getDirections(start,end)

for step in steps:
    display(HTML(step['html_instructions']))
    
```
